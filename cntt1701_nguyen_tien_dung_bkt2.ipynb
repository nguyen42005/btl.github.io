```python
import pandas as pd

# Đường dẫn tới file Excel
file_path = './CNTT17-01-nguyễn tiến dũng-BKT2.xlsx'

try:
    # Đọc dữ liệu từ file Excel vào DataFrame
    df = pd.read_excel(file_path)

    # In dữ liệu ban đầu
    print("Dữ liệu ban đầu:")
    print(df.head())

    # Kiểm tra và xử lý dữ liệu thiếu
    df.fillna("Không rõ", inplace=True)

    # Kiểm tra và sửa dữ liệu sai định dạng
    df["Số lượng"] = pd.to_numeric(df["Số lượng"], errors='coerce').fillna(0).astype(int)
    df["Giá trị đơn hàng"] = pd.to_numeric(df["Giá trị đơn hàng"], errors='coerce').fillna(0)
    df["Giá trị giảm giá"] = pd.to_numeric(df["Giá trị giảm giá"], errors='coerce').fillna(0)
    df["Giá trị thanh toán sau giảm giá"] = pd.to_numeric(df["Giá trị thanh toán sau giảm giá"], errors='coerce').fillna(0)

    # Loại bỏ các dòng có giá trị không hợp lệ (ví dụ: giá trị âm)
    df = df[(df["Số lượng"] > 0) & (df["Giá trị đơn hàng"] > 0)]

    # In dữ liệu sau khi làm sạch
    print("\nDữ liệu sau khi làm sạch:")
    print(df.head())

    # Lưu file mới
    new_file_path = "CNTT17-01-nguyễn tiến dũng-BKT2_cleaned.xlsx"
    df.to_excel(new_file_path, index=False)

    print(f"Dữ liệu đã được làm sạch và lưu vào {new_file_path}")

except FileNotFoundError:
    print(f"Không tìm thấy file Excel tại đường dẫn: {file_path}")
except Exception as e:
    print(f"Đã xảy ra lỗi: {e}")
```

## Data cleaning

### Subtask:
Clean the data in the DataFrame `df` by handling potential issues such as incorrect data types, missing values, and outliers.


**Reasoning**:
I will perform the data cleaning steps as described in the instructions. This includes checking and correcting data types, handling missing values (although there are none), checking for and handling outliers, and removing duplicate rows. After performing these actions, I will print the shape, information, and the first 5 rows of the DataFrame to verify the cleaning.



```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Check and correct data types
# From the data exploration, we know 'Ngày giao dịch' should be datetime
df['Ngày giao dịch'] = pd.to_datetime(df['Ngày giao dịch'], format='%d/%m/%Y')

# 2. Check for missing values (already checked in the exploration, but good practice)
# No missing values were found in the data exploration step, so no action is needed.

# 3. Identify potential outliers in numerical columns
numerical_cols = ['Số lượng', 'Giá trị đơn hàng', 'Giá trị giảm giá', 'Giá trị thanh toán sau giảm giá']

# Create box plots to visualize outliers
plt.figure(figsize=(12, 6))
for i, col in enumerate(numerical_cols):
    plt.subplot(2, 2, i+1)
    sns.boxplot(x=df[col])
    plt.title(f'Box Plot of {col}')
plt.tight_layout()
plt.show()

# Calculate IQR and identify outliers
for col in numerical_cols:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    # Remove outliers that are outside the bounds
    df = df[(df[col] >= lower_bound) & (df[col] <= upper_bound)]

# 5. Check for and remove duplicate rows
df.drop_duplicates(inplace=True)

# 6. Print the shape of the DataFrame
print("Shape of the DataFrame after cleaning:", df.shape)

# 7. Print the information of the DataFrame
print("\nInformation of the DataFrame after cleaning:")
df.info()

# 8. Print the first 5 rows of the DataFrame
print("\nFirst 5 rows of the DataFrame after cleaning:")
display(df.head())
```
